# 1: Check For Prime Number

```python
A = int(input("Enter a number: "))                    #Take input from user and store in A , Ex: A = 5
prime = True                                          #Consider the user number is prime 
if A <= 1:                                            # Any number less than 1 is not a prime number
    prime = False
else:
    i = 2                     #Start checking from 2 becoz 1 is neither prime nor not prime
    while i < A:              #create a while loop becoz if the condition is true so loop will execute 2<5? True !!
        if A % i == 0:        #give condition 5 % 2==0? No, condition is false
            prime = False
            break
        i += 1   #this increment is used to break the while loop 
if prime:
    print(f"{A} is a prime number.")   #Print the 5 is a prime number becoz 5%2 is not 0 
else:
    print(f"{A} is not a prime number.")
```

# 4: Word counter

```python
import re                                           # Regular expressions module for cleaning text
text = input("Enter a sentence or paragraph: ")     # Ask the user to enter text
text = text.lower()                                 # Convert text to lowercase
text = re.sub(r'[^a-z0-9\s]', '', text)             # Use re to remove all punctuation and special characters (keep only words and spaces)
words = text.split()                                # Split text into individual words
word_count = {}                                     # Create an empty dictionary to store word counts
for word in words:                                  # Count each word manually
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("\nWord Count:")                              # Print the result
for word in word_count:
    print(f"'{word}': {word_count[word]}")
```

# 5: Check Palindrome using is_palindrome function

```python
import re
def is_palindrome(text):  # Convert text to lowercase coz A!=a
    text = text.lower()
    cleaned = re.sub(r'[^a-z0-9]', '', text)    # Remove all non-alphanumeric characters (like punctuation and spaces)
    return cleaned == cleaned[::-1]             # Check if cleaned text is equal to its reverse
user_input = input("Enter a word, phrase, or number: ")  # Ask the user to enter some text
print(is_palindrome(user_input))     # Call the function and print the result (True or False)
```
